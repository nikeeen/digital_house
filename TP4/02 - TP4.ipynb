# 1. Introducción

##### El objetivo del presente trabajo consiste en predecir la demanda de energía en los próximos cinco días en Argentina. Para ésto contamos con un dataset extraído de la pagína de CAMMESA (Compañía Administradora del Mercado Mayorista Eléctrico S.A) [Histórico valores energia y potencia](http://cammesaweb.cammesa.com/2022/06/24/maximos-historicos-de-energia-y-potencia-estacionales/). Consiste en una serie histórica desde enero de 2007 a mayo 2022 de la energía eléctrica distribuída diariamente a todo el país. El dataset cuenta con las siguientes features:

- AÑO: corresponde al año de dato
*MES: mes del dato
*N° MES: corresponde al número de mes (1 a 12 - enero a diciembre)
*VERANO/INVIERNO: corresponde a la estación del año (solo separa en verano e invierno)
*SEMANA: número de semana desde enero 07 a mayo 22
*FECHA: día del dato
*TIPO DIA: se considera días hábiles (laborables), feriados, sábados y domingos.
*DIA: corresponde al día de la semana
*N° DIA: corresponde al número de día del dato
*Energia SADI (GWh): consumos total de energía*
*Potencia Pico SADI (MW): potencia máxima del día
*Hora Potencia Pico: hora en que se produjo la potencia máxima del día.
*Temperatura Media Diaria GBA (°C): promedio de temperatura del día en Gran Buenos Aires.
*Estado del Tiempo: variable cualitativa donde indica si el día estuvo claro, nublado, o seminublado.


##### *Energía SADI: El Sistema Argentino de Interconexión (SADI) es una red eléctrica conformada por tendidos de alta tensión que interconectan las distintas regiones de Argentina. Es la red que recolecta y transporta toda la energía eléctrica que se genera en el país. La longitud del SADI es de 20.296 kilómetros de distribución troncal, y 14.197 km de líneas de 500 kV. Su objetivo es que todo el país esté conectado por un sistema eléctrico común y único.

## 4. Modelo: Regresión lineal con mínimos cuadrados
[volver al Indice](#section_toc)

#### Como primer modelo vamos a aplicar un modelo de regresión lineal con mínimos cuadrados clásicos, para ésto vamos a tener que generar nuevas variables que son requeridas por este tipo de modelos. Los modelos de regresión lineal no pueden trabajar sobre variables de tipo fecha, para esto utilizaremos una variable dummy de tiempo donde a cada fecha se le asignará un valor, esto nos va a permitir hace regresiones contra el tiempo y estimar la tendencia. Luego convertiremos las variables categóricas en variables dummies: 'tipo_dia', 'dia', 'estado_tiempo', y 'nueva_estacion'. 'Estación' no la consideramos porque generamos una nueva columna ('nueva_estacion') donde se incorporaron las estaciones otoño y primavera. Todas estas variables son las que van a estar relacionadas con la estacionalidad.


```python
#Generamos la variable 'timeindex' como variable dummy de tiempo
data['timeindex'] = pd.Series(np.arange(len(data.energia_sadi)), index=data.index)

#Corroboramos que la columna timeindex se haya generado correctamente, vemos el inicio del dataframe y el final.
data.head()
```

```python
print(data.shape)
data.tail()
```

```python
#Vamos a convertir los meses en dummies con el método get_dummies
#Aplicamos drop_first= True para envitar la colinealidad entre variables.

#las siguientes dos fueron agregadas porque en linux no me toma el integer de una, sino que las toma como float
#comentar en el caso que no sean necesarias
data['numero_mes'] = data['numero_mes'].astype(int)
data['numero_mes'] = data['numero_mes'].astype(str)


data_dummie_mes =pd.get_dummies(data['numero_mes'], prefix='mes', drop_first=True)   


#Vamos a convertir el resto de las variables categóricas en dummies.
data_dummies=pd.get_dummies(data[['tipo_dia', 'dia', 'estado_tiempo', 'nueva_estacion']], drop_first=True )

#Unimos los dos dataframes de dummies y el que teniamos originalmente
data_dm=data_dummie_mes.join(data_dummies)
data=data.join(data_dm)
data.head()
```

```python
print("Continuamos teniendo la misma cantidad de filas:", data.shape[0])
print("Cantidad de columnas nueva:", data.shape[1])
print("Cantidad de columnas anteriores 16")
print("-----------------------------------------------")
print("Las nuevas columnas ahora son:")
print("                                               ")
print(data.columns)
```

#### Vamos a generar una función para calcular el error cuadrático medio (RMSE) para comparar los valores predichos con los observados y así obtener una medida de performace.

```python
def RMSE(predichos, observados):
    mse = (predichos - observados) ** 2
    rmse = np.sqrt(mse.sum() / mse.count())
    return rmse
```

#### Realizamos cross validation para series de tiempo para no perder las dependencias temporales entre las observaciones.

```python
#Aplicamos un split entre train y test para series de tiempo
tscv = TimeSeriesSplit(n_splits=5)
```

```python
#Definimos una función que aplica time series cross validation para el modelo single exponential smoothing
def timeseriesCVscore_exp_smoot(alpha, series):
    """
        Returns error on CV  
        
        params - vector of parameters for optimization
        series - dataset with timeseries
        slen - season length for Holt-Winters model
    """
    # creamos un array de errores:
    errors = []
    
    values = series.values
    
    # instanciamos el objeto que realiza el tscv:
    tscv = TimeSeriesSplit(n_splits=5) 
    
    # Aplicamos cross validation:

    for train, test in tscv.split(values):
    
        model = SimpleExpSmoothing(values[train]).fit(smoothing_level=alpha,\
                                                             optimized=False)
        
        predictions = model.forecast(len(test))
        actual = values[test]
    
        error = mean_squared_error(predictions, actual)
        errors.append(error)
        
    return np.mean(np.array(errors))
```

```python
alphas = [0.001, 0.01, 0.1, 0.2, 0.3, 0.35, 0.4, 0.5, 0.7]
errors = []

for alpha in alphas:
    error = timeseriesCVscore_exp_smoot(alpha, data_train.energia_sadi)
    errors.append(error)

print('Alpha óptimo:', alphas[np.argmin(errors)])
```

```python
# Entrenamos el modelo optimizado:

model_exp_smoothing = SimpleExpSmoothing(data_train.energia_sadi).fit(smoothing_level=alphas[np.argmin(errors)],\
                                                             optimized=False)
```

```python
#Generamos una columna nueva con las predicciones para 5 días
data_test["Simple_Smoothing"] = model_exp_smoothing.forecast(5)
```

```python
#Calculamos el RMSE del modelo
print("--------------------------------------------------------")
print("RMSE Simple Exponential Smoothing", RMSE(data_test["Simple_Smoothing"], data_test.energia_sadi))
print("--------------------------------------------------------")
```

#### Para poder aplicar ARIMA primero debemos testear la estacionalidad de la serie

```python
# Testeamos la estacionariedad de los residuos:

res_model = data['energia_sadi']

result = adfuller(res_model)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
for key, value in  result[4].items():
    print('Valor crítico %s: %.2f' % (key,value))
```

#### El resultado del test de Dickey-Fuller nos da un p-value < 0.05, lo que significa que no rechazo la hipotesis nula de que la serie es estacionada, por lo tanto puedo aplicar directamente el modelo de ARIMA.

```python
def tsplot(y, lags=None, figsize=(12, 7), style='bmh'):
    """ 
        Plotea la serie de tiempo, el ACF y PACF y el test de Dickey–Fuller
        
        y - serie de tiempo
        lags - cuántos lags incluir para el cálculo de la ACF y PACF
        
    """
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
        
    with plt.style.context(style):    
        fig = plt.figure(figsize=figsize)
        layout = (2, 2)
        
        # definimos ejes
        ts_ax = plt.subplot2grid(layout, (0, 0), colspan=2)
        acf_ax = plt.subplot2grid(layout, (1, 0))
        pacf_ax = plt.subplot2grid(layout, (1, 1))
        
        y.plot(ax=ts_ax)
        
        # obtengo el p-value con h0: raiz unitaria presente
        p_value = sm.tsa.stattools.adfuller(y)[1]
        
        ts_ax.set_title('Análisis de la Serie de Tiempo\n Dickey-Fuller: p={0:.5f}'\
                        .format(p_value))
        
        # plot de autocorrelacion
        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax)
        # plot de autocorrelacion parcial
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax)
        plt.tight_layout()
```

```python
tsplot(res_model, lags=90)
```

#### NOTA: Por una cuestión de simplicidad y poca diferencia con el ARIMA, sólo se va a realizar con el modelo base de XGBoost

```python
def modelo_dias(data, cantidad_de_dias):

    data_train, data_test = train_test_split(data, test_size=cantidad_de_dias, random_state=123, shuffle=False)

    #Separamos en train y test para que solo tome las features necesarias.

    X_train = data_train.drop(['año', 'mes', 'numero_mes', 'estacion', 'semana', 'date', 'tipo_dia',
        'dia', 'numero_dia', 'energia_sadi', 'potencia_pico', 'hora_pico','estado_tiempo', 'nueva_estacion','tipo_dia_FERIADO','dia_Jueves', 'dia_Lunes', 'dia_Martes',
        'dia_Miércoles', 'dia_Sábado', 'dia_Viernes'], axis=1)
    y_train=data_train.energia_sadi

    X_test = data_test.drop(['año', 'mes', 'numero_mes', 'estacion', 'semana', 'date', 'tipo_dia',
        'dia', 'numero_dia', 'energia_sadi', 'potencia_pico', 'hora_pico','estado_tiempo', 'nueva_estacion','tipo_dia_FERIADO','dia_Jueves', 'dia_Lunes', 'dia_Martes',
        'dia_Miércoles', 'dia_Sábado', 'dia_Viernes'], axis=1)
    y_test=data_test.energia_sadi
    
    #Instanciamos el modelo
    xgb_regressor = XGBRegressor()

    #Entrenamos el modelo
    xgb_regressor = xgb_regressor.fit(X_train,y_train)

    data_train['predict_xgb'] = xgb_regressor.predict(X_train)
    data_test['predict_xgb'] = xgb_regressor.predict(X_test)
    
    print("El RMSE de train es: ", RMSE(data_train.predict_xgb, data_train.energia_sadi))
    
    #Calculamos el RMSE de Test
    print("El RMSE de test es: ", RMSE(data_test.predict_xgb, data_test.energia_sadi))
    
    plot = data_test.plot(kind="line", y = ["energia_sadi", "predict_xgb"], figsize=(8,3));    
```
