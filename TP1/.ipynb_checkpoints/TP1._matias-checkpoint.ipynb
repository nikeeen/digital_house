```python
# C- TOMO EL DATO CAPTURADO QUE YA ES NUMERO DE LA VARIABLE Y LUEGO, PARA LAS VARIABLES CON DATOS DE NOMBRE DE NUMEROS, APLICO LA FUNCION QUE CONVIERTE LOS NOMBRES DE NUMEROS A NUMEROS  

df['description_amb_num1'] = pd.to_numeric(df['description_amb'], errors='coerce') 
df['description_amb_num2'] = pd.to_numeric(df['description_amb_2'], errors='coerce')
df['description_amb_text_num'] = df['description_amb'].apply(digit_str_to_int)

# AQUI MEZCLO TITLE Y DESCRIPTION CON VALORES A UN ESPACIO DE DISTANCIA DE LA PALABRA CLAVE PARA ARMAR UNA SOLA VARIABLE
df['description_cuartos_num'] = pd.to_numeric(df['description_cuartos'], errors='coerce')
df['title_cuartos_num'] = pd.to_numeric(df['title_cuartos'], errors='coerce')
df['description_cuartos_text_num'] = df['description_cuartos'].apply(digit_str_to_int)
df['title_cuartos_text_num'] = df['title_cuartos'].apply(digit_str_to_int)

df['description_m2_num'] = pd.to_numeric(df['description_m2'], errors='coerce')
df['title_m2_num'] = pd.to_numeric(df['title_m2'], errors='coerce')
df['description_m2_text_num'] = df['description_m2'].apply(digit_str_to_int)
df['title_m2_text_num'] = df['title_m2'].apply(digit_str_to_int)

# print("description_amb_num1: ",df['description_amb_num1'].value_counts())
# print("description_amb_num2: ",df['description_amb_num2'].value_counts())
# print("description_amb_text_num: ",df['description_amb_text_num'].value_counts())
```

```python
# 2- TRABAJO SOBRE LAS VARIABLES DE TEXTO DESCRIPTIVAS 

# D- CREO UNA FUNCION PARA RECODIFICAR LA VARIABLE DESCRIPTION_AMB CON LOS VALORES DE LAS CREADAS ANTERIORMENTE  
#  ESTA FUNCION SOLO TIENE 3 PARAMETROS Y RESPETA UNA JERARQUIA DE RECODIFICACION:

def recode_func_amb(a,b,c):
    if a>0:
        return a
    if b>0:
        return b
    if c>0:
        return c
    else:
        return a

# APLICO LA FUNCION EN LA VARIABLE FINAL DE AMBIENTES    
df['cantidad_ambientes'] = df.apply(lambda x: recode_func_amb(x['description_amb_num1'], x['description_amb_num2'],x['description_amb_text_num']), axis=1)

# COMPRUEBO QUE TODO HAYA FUNCIONADO BIEN:
df2=df.loc[(df['description_amb_num2'] > 0) | (df['description_amb_text_num'] > 0 )]

print("Ejemplo para ver si recodifica bien con variable a: ", df.loc[:,['cantidad_ambientes','description_amb_num1', 'description_amb_num2','description_amb_text_num']].head(9))
print("Ejemplo para ver si recodifica bien con variables b y c: ", df2.loc[:,['cantidad_ambientes','description_amb_num1', 'description_amb_num2','description_amb_text_num']].head(9))
```

```python
# 2- TRABAJO SOBRE LAS VARIABLES DE TEXTO DESCRIPTIVAS 

# D- CREO UNA FUNCION PARA RECODIFICAR LAS VARIABLES QUE CAPTURAN EL DATO DE CUARTO (CANTIDAD DE HABITACIONES) Y METROS CUADRADOS  
#  ESTA FUNCION NECESITA 4 PARAMETROS Y RESPETA UNA JERARQUIA DE RECODIFICACION:

def recode_func_4param(a,b,c,d):
    if a>0:
        return a
    if b>0:
        return b
    if c>0:
        return c
    if d>0:
        return d
    else:
        return a

# APLICO LA FUNCION EN LA VARIABLE FINAL DE CUARTOS    
df['cantidad_cuartos'] = df.apply(lambda x: recode_func_4param(x['description_cuartos_num'], x['title_cuartos_num'],x['description_cuartos_text_num'],x['title_cuartos_text_num']), axis=1)

# APLICO LA FUNCION EN LA VARIABLE FINAL DE METROS CUADRADOS
df['cantidad_m2'] = df.apply(lambda x: recode_func_4param(x['description_m2_num'], x['title_m2_num'],x['description_m2_text_num'],x['title_m2_text_num']), axis=1)

# COMPRUEBO QUE TODO HAYA FUNCIONADO BIEN:
df3=df.loc[(df['title_cuartos_num'] > 0) | (df['description_cuartos_text_num'] > 0 )]
df4=df.loc[(df['title_m2_num'] > 0) | (df['description_m2_text_num'] > 0 )]

print("CUARTOS ver si recodifica bien con variable a: ", df.loc[:,['cantidad_cuartos','description_cuartos_num','title_cuartos_num', 'description_cuartos_text_num','title_cuartos_text_num']].head(9))
print("CUAROTS ver si recodifica bien con variables b y c: ", df3.loc[:,['cantidad_cuartos','description_cuartos_num','title_cuartos_num', 'description_cuartos_text_num','title_cuartos_text_num']].head(9))

print("M2 ver si recodifica bien con variable a: ", df.loc[:,['cantidad_m2','description_m2_num','title_m2_num', 'description_m2_text_num','title_m2_text_num']].head(9))
print("M2 ver si recodifica bien con variables b y c: ", df3.loc[:,['cantidad_m2','description_m2_num','title_m2_num', 'description_m2_text_num','title_m2_text_num']].head(9))
```
